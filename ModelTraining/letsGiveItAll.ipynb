```python
import pandas as pd

# Load CSV files
catering_df = pd.read_csv('tbl_catering.csv')
catering_details_df = pd.read_csv('tbl_catering_details.csv')
catering_products_df = pd.read_csv('tbl_catering_products.csv')

# Merge Catering Data with Catering Details
merged_data = pd.merge(catering_data, catering_details, left_on='ID', right_on='CATERING_ID', suffixes=('_data', '_details'))

# Rename columns to maintain consistency
merged_data.rename(columns={
    'ITEM_DESC_details': 'ITEM_DESC',
    'PRICE_details': 'PRICE_details'  # To avoid confusion with the `PRICE` column
}, inplace=True)

# Ensure 'ITEM_DESC' and 'PRODUCT_NAME' exist before the second merge
if 'ITEM_DESC' not in merged_data.columns:
    raise KeyError("Column 'ITEM_DESC' not found in merged_data")
if 'PRODUCT_NAME' not in catering_products.columns:
    raise KeyError("Column 'PRODUCT_NAME' not found in catering_products")

# Merge with Catering Products
merged_data = pd.merge(merged_data, catering_products, left_on='ITEM_DESC', right_on='PRODUCT_NAME', how='left')

# Inspect the merged data
print("Columns after second merge:", merged_data.columns)
print(merged_data.head())
```
